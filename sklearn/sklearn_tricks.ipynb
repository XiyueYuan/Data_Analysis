```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Right-skewed and Left-skewed

- For data that is right-skewed, we use log to transform the data, in numpy, `log1p()` to log(1 + x) for all values in data. Why 1 + x? Because some values may by 0.
- $logx' = \frac{1}{x}$, so when x is larger, the increasing rate gets slower 

```python
prices = [
    120000, 130000, 125000, 140000, 160000, 155000, 170000, 180000, 175000,
    200000, 210000, 220000, 250000, 270000, 300000,
    350000, 400000, 450000, 500000, 600000, 750000, 900000,
    1200000, 1600000, 2500000, 3500000
]
prices_log = np.log1p(prices)
plt.hist(prices, bins = 10)
plt.show()
```

```python
plt.hist(prices_log, bins = 10)
plt.show()
```

- when data is left-skewed, we use the opposite of $logx$, we give each value in the list a square of themselve
- after squaring, the value on the right(the large ones) will be stretched or spread out so that the overall distribution looks less dense

```python
left_skewed = [
    1, 2, 3, 4, 5, 
    10, 20, 30, 40, 50,
    60, 65, 70, 72, 73, 74, 75, 76, 77, 78,
    79, 80, 81, 82, 83, 84, 85, 86, 87, 88,
    89, 90, 91, 92, 93, 94, 95, 96, 97, 98, 99, 100
]
left_exp = np.array(left_skewed) ** 2
plt.hist(left_skewed, bins = 10)
plt.show()
```

```python
plt.hist(left_exp, bins = 10)
plt.show()
```
