```python
import pandas as pd
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col = 'ResponseId')
df.head(10)
```

To get the specific column after filtering
- first create a variable contains the selected dataframe
- then put into loc to filter

```python
high_salary = df['ConvertedCompYearly'] >= 10000000
df.loc[high_salary, ['LearnCodeChoose', 'LearnCode', 'Age']]
```

Using `isin` as `in` function to filter

```python
countries = ['United States', 'India', 'United Kindom', 'Germany']
filt = df['Country'].isin(countries)
df.loc[filt, 'Country']
```

`str.contains` will check if each `'LanguageHaveWorkedWith'` contains `str` `'Python'` and return a bool value

`na = False` by default

```python
filt = df['LanguageHaveWorkedWith'].str.contains('Python', na = False)
df.loc[filt, 'LanguageHaveWorkedWith']
```

Rename the column name by using `{}`

```python
df = df.rename(columns = {'LanguageHaveWorkedWith': 'LanguageUsed'})
df['LanguageUsed']
```

```python
df['Employment']
```

Now, if i want to convert all `'Employed'` to `True`, use the `repr()` to check the exact str in it, because it might be `'Employed '` with an extra space after, and use `type()` to check the type, because it might be non-str elements.

```python
print(type(df['Employment'].iloc[1]))
print(repr(df['Employment'].iloc[1:10]))
```

Then use `map(lambda x:)` to modify every element in the `Employment` column list

```python
df['Employment'] = df['Employment'].map(lambda x: True if x == 'Employed' else False)
df['Employment']
```

```python
df['Country']
```

Using `sort` function to sort people salary within each country


```python
df = df.rename(columns = {'ConvertedCompYearly': 'Salary'})
df['Salary']
```

```python
test = df.sort_values(by = ['Country', 'Salary'], ascending = [True, True])
test[['Salary', 'Country']].head(10)
```

`nlargest(int)` filt the largest one then using `loc[filt.index]` select other columns

```python
filt = df['Salary'].nlargest(10)
df.loc[filt.index, 'Country']
```

Or you can simply display all, or `nsmallest`

```python
df.nlargest(10, 'Salary')
```
