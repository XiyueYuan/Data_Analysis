If you want multiple lines in the same plot graph, put all `plt.plot()` before the `plt.show()`

`plt.legend` is used to label the lines in order if you write them separately, or you can write `plt.plot(dev_x, dev_y, label = 'Data1')`

In label, specify the color and line pattern, format strings refer to [', 'https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html''](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

All customized: `plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)`

```python
plt.plot(dev_x, dev_y, 'b-', marker = '^')
plt.plot(dev_x, py_dev_y, 'g--',linewidth = 3, marker = 'o')
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary by age')
plt.grid(True)
plt.style.use('fivethirtyeight')
plt.legend(['Data1', 'Data2'])
plt.show()
```
