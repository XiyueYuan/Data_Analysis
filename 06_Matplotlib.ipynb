## Histogram + Scatter plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

# Here bins = 5 represents the number of bins is 5
plt.hist(ages, bins = 5, edgecolor = 'black')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.tight_layout()
```

```python
bins = [20, 30, 40, 50, 60]
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
# Or you can define bins by yourself 

plt.hist(ages, bins = bins, edgecolor = 'black')
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('assets/data/respondent.csv')
print(data.head())
print(data.shape)
ids = data['Responder_id']
age = data['Age']
```

```python
bins = [10 * x for x in range(1, 11)]
plt.hist(age, bins = bins, edgecolor = 'black', log = True)
plt.title('Ages of Respondents')
median_age = age.median()
plt.axvline(median_age, color = 'red', label = 'Age Median', linewidth = 2)
plt.legend()
plt.show()
```

```python
import random
plt.style.use('ggplot')
x = [random.randint(1, 10) for _ in range(20)]
y = [random.randint(1, 10) for _ in range(20)]
plt.scatter(x, y, s = 80, c = 'green', marker = 'x') # s represents dot size, c represents dot color 
plt.tight_layout()
plt.show()
```

### Plotting Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
df = pd.read_csv('assets/data/time_data.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'])
print(df.columns)
print(df['Volume'])
```

```python
date = df['Date']
volume = df['Volume']
plt.style.use('fivethirtyeight')
plt.plot_date(date, volume, linestyle = 'solid')
plt.tight_layout()

# Date Format in matplotlib 
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.show()
```
